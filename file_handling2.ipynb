```python
# reading a  binary data

with open("image.jpg","rb") as f:
    data=f.read()
    print(data)
    print(type(data))
    print('size:',len(data),"bytes")
    print("First 4 bytes ", data[:4])
```

```python
## copying a binary file 

with open("image.jpg","rb") as read:
    with open("new_image.jpg","wb") as new:
        new.write(read.read())
```

```python
##  Reading files--Modes
# 1---> f.read() 
# 2---> f.read(n)
# 3---> f.readline()
```

```python
with open("poem.txt","w") as f:
    f.write('Roses are red \n')
    f.write('Violets are blue \n')
    f.write('python is great \n')
    f.write('And so are you \n')
```

```python
with open("poem.txt","r") as f:
    print(f.read())
```

```python
with open("poem.txt","r") as f:
    first6=f.read(6)
    print(first6)
    next5=f.read(5)
    print(next5)
```

```python
# Readline()

with open("poem.txt","r") as f:
    line1=f.readline()
    print(line1)
    Line2=f.readline()
    print(Line2)
    Line3=f.readline()
    print(Line3)
    Line4=f.readline()
    print(Line4)
    Line5=f.readline()
    print(Line5)

```

```python
## readlines()- All lines into a list.

with open("poem.txt","r") as f:
    lines=f.readlines()
    print(lines)
    print(type(lines))
```

```python
# Acessing specific lines by index

print("First line :", lines[0])
print("Last One:",lines[-1])
```

```python
# sort line Alphabetically #

sorted_lines = sorted(lines)
print(sorted_lines)
```

```python
sort_by_lambda= sorted(sorted_lines,reverse=True)
print(sort_by_lambda)
```

```python
for line in sorted_lines:
    print(line)
```
