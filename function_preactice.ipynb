```python
# calculater #

num1=int(input("Enter Your Number"))
num2=int(input("Enter Your Number"))

def calulater(num1,num2):
    
    Add=num1+num2
    print(f"Addition:- {num1}+{num2} = {Add}")
    subtract=num1-num2
    print(f"Subtraction:- {num1}-{num2} = {subtract}")
    multiply=num1*num2
    print(f"multiplication:- {num1}*{num2} = {multiply}")
    Average=(num1+num2)/2
    print(f"Average is = {Average}")

calulater(num1,num2)
```

```python
## types of function Arguments ## 

# 1--> required arguments (Single/multiple arguments)
# 2--> Default argument
# 3--> Keyword arguments (name arguments)
# 4--> Arbitary arguments (Variable length arguments *args and **kwargs)
```

```python
## Required Arguments  (single/ multiple arguments)##

def grettings(name): # name is a Parameters #
    print(f"Hello {name}")

grettings("Abhishek") # Abhishek is Arguments #
```

```python
def intro(course_name,instructor_name):
    print(f"Welcome to {course_name} Course by {instructor_name}")

intro("python","Lohit Sir")
```

```python
## Default Argument ##

def gretting(name="World"): # World is A default value #
    print(f"Hello {name}")

gretting()  
gretting("Abhishek")
```

```python
## keyword Arguments ##

def divide(a,b):
    return a/b

print(divide(100,10))

result=divide(b=10,a=200)

print(result)
```

```python
## Arbitary postional Arguments (*agrs) ##
# arguments are store as a tuple #

def add_number(*number):
    return sum(number)

add_number(1,2,4,5,7)
```

```python
def even_odd(*numbers):

    for num in numbers:
        if num%2==0:
            print(f"{num} : Even")
        else:
            print(f"{num} : Odd")

even_odd(10,12,4,7,8,9,1,4,5,6,2,3,7,5)
```

```python
def greet(*names):
    for name in names:
        print(f"Hello {name}")

greet("Abhishek","Anish","Ankit","Anil","Ankhus","Adrash","Ankul")
```

```python
## Arbitary keyword Arguments (**Kwargs) ##

## the arguments are stored as a dictionary. ##

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")

print_details(name="Abhishek",age=21,city="Kolkata")
```
