```python
fig,ax = plt.subplots(2,1,sharex=True) ## sharex matlab x axis dono ka same rahega akag alag nhi ayega   ## yaha pe hum 2,1 diye hain iska matlab hua 2 subplots ko hum log plot kar sakte hain 1 ka matlab 1 colloumn main 

ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title("avg vs strike rate")
ax[0].set_ylabel("strike Rate")

ax[1].set_title("avg vs runs")
ax[1].set_ylabel("runs")
ax[1].set_xlabel("Avg")
```

```python
fig,ax = plt.subplots(2,1,figsize=(16,8)) ## sharex matlab x axis dono ka same rahega akag alag nhi ayega   ## yaha pe hum 2,1 diye hain iska matlab hua 2 subplots ko hum log plot kar sakte hain 1 ka matlab 1 colloumn main 

ax[0].scatter(batter['avg'],batter['strike_rate'],color="red")
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title("avg vs strike rate")
ax[0].set_ylabel("strike Rate")

ax[1].set_title("avg vs runs")
ax[1].set_ylabel("runs")
ax[1].set_xlabel("Avg")
```

```python
fig,ax = plt.subplots(2,2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'])   ## ab ye 2D array ban chuka hain toh hume 2 value provide karna parega ax main
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
fig.show()
```

```python
## second way for subplots ##

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)  ## 2,1,1 meaning 2 rows 1 column aur data 1 
 
ax1.scatter(batter['avg'],batter['runs']) 

ax2 = fig.add_subplot(2,1,2)

ax2.hist(batter['runs'])
```

```python
## 3D surface plots

x = np.linspace(-10,10,100) 

y = np.linspace(-10,10,100) 
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2

z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
## contour plot ## 3D plot ko 2D main convert karta hain 

fig = plt.figure(figsize=(12,8)) 

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
## countuor f 

fig = plt.figure(figsize=(12,8)) 

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
