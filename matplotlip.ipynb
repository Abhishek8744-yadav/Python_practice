```python
plt.plot(batsman["index"],batsman['V Kohli'],color='green',marker="D",markersize=10)
plt.plot(batsman["index"],batsman['RG Sharma'],color='blue',marker="D")

plt.title("Rohit sharma vs Virat kholi Career Comparision")

plt.xlabel("Season")

plt.ylabel("Runs Scored")
```

```python
## giveing name to the line ## 

plt.plot(batsman["index"],batsman['V Kohli'],color='green',marker="D",label='Virat')
plt.plot(batsman["index"],batsman['RG Sharma'],color='blue',marker="D",label ="Rohit")

plt.title("Rohit sharma vs Virat kholi Career Comparision")

plt.xlabel("Season")

plt.ylabel("Runs Scored")
plt.legend()   ## ----> iss function ko call karoge toh hi naam show hoga 
```

```python
batsman_season
```

```python
plt.bar(batsman_season["batsman"],batsman_season["2015"])
```

```python
plt.bar(batsman_season["batsman"],batsman_season["2015"])
plt.bar(batsman_season["batsman"],batsman_season["2016"])

## ye graph ye bata raha hain ki jo 2015 main baki log run banaya tha usse jayda 2016 main banyaya hain par ms dhoni se 2016 main 2015 se kam run banya hain iss liye 
## ms dhoni ka naya wala bar orage wala nhi cover kar paya bule wale ko baki sab ka cover ho gaya hain purane wale bar k sath 
```

```python
plt.bar(batsman_season["batsman"],batsman_season["2015"],width=.5)
```

```python
plt.bar(np.arange(batsman_season.shape[0]) - 0.2,batsman_season['2015'],width=0.2)
plt.bar(np.arange(batsman_season.shape[0]),batsman_season['2016'],width=0.2)
plt.bar(np.arange(batsman_season.shape[0]) + 0.2,batsman_season['2017'],width=0.2)
```

```python
plt.bar(np.arange(batsman_season.shape[0])-0.2,batsman_season["2015"],width=0.2,color='yellow')
plt.bar(np.arange(batsman_season.shape[0]),batsman_season["2016"],width=0.2,color='red')
plt.bar(np.arange(batsman_season.shape[0])+0.2,batsman_season["2017"],width=0.2,color='blue')

plt.xticks(np.arange(batsman_season.shape[0]),batsman_season["batsman"])   ## xticks replace kar diya shapes wale number ko batsman k naam se 

plt.show
```

```python
## managing big names ##

children =[10,20,40,10,30]

colors =["red red red red red","blue blue blue blue blue blue blue","green green green green green green"," yellow yellow yellow yellow yellow","pink pink pink pink pink pink"]

plt.bar(colors,children)

plt.xticks(rotation='vertical')

plt.show
```

```python
## stacked bar chat ## 

batsman_season

plt.bar(batsman_season["batsman"],batsman_season["2015"],label="2015")

plt.bar(batsman_season["batsman"],batsman_season["2016"],bottom=batsman_season["2015"],label="2016")

plt.bar(batsman_season["batsman"],batsman_season["2017"],bottom=batsman_season["2016"] + batsman_season["2015"],label="2017")

plt.legend()

plt.show()
```

```python
## PIE CHART ##

## univariate / bivariate Analysis

## Categorical vs numerical

## use case - to find contribution on a standard scale (100%)
```

```python
data =[23,45,100,20,49]  ## ye sare number ka sum hoge fir percentage nikalta hain jaise pura sum hone k baad 23 iss data main kitna percentage deta hain ##
subject = ["eng","science","math","evs","geo"]
plt.pie(data,labels=subject)
plt.show
```

```python
gayle = pd.read_csv("C:\\Users\\DAYA RAM YADAV\\Downloads\\gayle-175.csv")
```

```python
gayle
```

```python
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct='%0.1f%%')  ##autopct='%0.1f%%'   ye percentage show karta hain pie chart main 
plt.show
```

```python
## percentage and colors ## 

plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct='%0.1f%%',colors=['blue',"green","yellow","pink","cyan","brown"])
plt.show
```

```python
## explode shadow ##
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)  ## shadow kaam hain shadow dikhanan 
plt.show
```

```python
## save picture ##

## explode shadow ##
plt.pie(gayle["batsman_runs"],labels=gayle["batsman"],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0],shadow=True)  ## shadow kaam hain shadow dikhanan 

plt.savefig("sample_pie.png")
```
