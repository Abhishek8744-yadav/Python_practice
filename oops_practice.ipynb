```python
## object oriented programing ## Data and methods are bunlded together 

class BankAccount:
    def __init__(self,account_number,account_holder,balance):
        self.account_number=account_number
        self.account_holder=account_holder
        self.balance=balance

## function inside a class called methods 

    def deposit(self,amount):
        self.balance+=amount
        print(f"Deposited: {amount} New balance: {self.balance}")

    def withdraw(self,amount):
        if self.balance>=amount:
            self.balance-=amount
            print(f"Amount withdraw is {amount} your current Balance is {self.balance}")
        else:
            print("insuffeint balance")

    def check_balance(self):
        print("Current balance ", self.balance)



    

```

```python
## function inside class are called as method 

## class car (display_info,accelerate,brake,)

class Car:
    def __init__(self,brand,model,color):
        self.brand=brand
        self.model=model
        self.color=color
        self.speed=0


    def display_info(self):
        print(f"Car brand is {self.brand} Model is {self.model}")
        print(f"Color: {self.color}")
        print(f"Current Speed: {self.speed}")


    def accelerate(self,increase):
        self.speed+=increase
        print(f"{self.brand} is now going at {self.speed} km/h")

    
    def brake(self, decresase):
        self.speed -= decresase
        if  self.speed<0:
            self.speed=0
        print(f"{self.brand} slowed down to {self.speed}")
    
    
```

```python
## Class Student ##


class Student:
    def __init__(self,name,roll_num,grade):
        self.name=name
        self.roll_num=roll_num
        self.grade=grade
        self.subject=[]
        self.marks={}

    
    def add_subject(self,subject_name):
        if subject_name not in self.subject:
            self.subject.append(subject_name)
            print(f"Subject {subject_name} Added for {self.name}")
        
        else:
            print(f"{self.name} is already enrolled in {subject_name}")


    def add_marks(self,subject_name,marks):
        if subject_name in self.subject:
            self.marks[subject_name]=marks
            print(f"Marks updated: {subject_name} = {marks}")
        
        else:
            print(f"Error : {subject_name} Not found in enrolled subjects")

    
    def calculate_average(self):
        if len(self.marks) == 0:
            return 0
        total = sum(self.marks.values())
        average=total/len(self.marks.values())
        return average
    

    def display_info(self):
        print(f"Student Name {self.name}")
        print(f"Student Roll number {self.roll_num}")
        print(f"Grade : {self.grade}")
        print(f"Subject enrolled: {','.join(self.subject) if self.subject else "None"} ")
        if self.marks:
            print("Marks:")
            for subject,marks in self.marks.items():
                print(f"{subject} : {marks}")
            print(f"Average: { self.calculate_average():.2f}")
        else:
            print("No Marks recorded yet")

    
    def Check_result(self):
        avg=self.calculate_average()
        if avg>=40:
            print(f"{self.name}  has Passed with {avg}%")

        else:
            print(f"{self.name} has failed with {avg}")
            
```
