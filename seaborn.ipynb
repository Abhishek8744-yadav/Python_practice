```python
## Scatter plot ## 

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
## Scatter plot ## ----> axes level function 

sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
## relplot --> figure level function --> square shape 

sns.relplot(data=tips,x='total_bill',y='tip',kind="scatter")
```

```python
## hue parameter 

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
## style parameter ##

sns.scatterplot(data=tips,x='total_bill',y='tip',hue="sex",style="time")
```

```python
## Size parameter ##
sns.scatterplot(data=tips,x='total_bill',y='tip',hue="sex",style="time",size="size")
```

```python
## line plot
gap = px.data.gapminder()

gap
```

```python
temp_df = gap[gap["country"] == "India"]
```

```python
## Axes level function  

sns.lineplot(data=temp_df,x="year",y="lifeExp")
```

```python
## Using relplot ##
sns.relplot(data=temp_df,x="year",y="lifeExp",kind="line")
```

```python
## Hue Style

temp_df = gap[gap["country"].isin(["India","Pakistan","China"])]
```

```python
sns.relplot(kind="line",data=temp_df,x="year",y="lifeExp",hue="country")  ## ye figure level hain isme legend figure k bahar rehta hain iska aces level figure level tak hain 
```

```python
sns.lineplot(data=temp_df,x="year",y="lifeExp",hue="country") ## ye axes level hain isme legeng ko axes k anadar hi rakhna parta hain kyuki uska acess axes level tak hi hain
```

```python
temp_df = gap[gap["country"].isin(["India","Brazil","Germany"])]
```

```python
sns.relplot(kind="line",data=temp_df,x="year",y="lifeExp",hue="country",style="continent",size="continent") 
```

```python
## facet plot --> its work only with figure level functions ---> its will work with relplot and it will not work with scatterplot and linelpot function 
tips

sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",col="sex")  ## col ka kaam hain sex coloumn main jitna v catagoery hain utne plot bana dega ##
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",row="sex")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="scatter",col="sex",row='smoker')
```

```python
## column wrap ##

sns.relplot(data=gap,x="lifeExp",y="gdpPercap",kind="scatter",col="year",col_wrap=4)  ## Column wrap = 4 ka matlab ye hota hain ki 1 row main 4 hi column hoga 
```

```python
## Distribution Plots ##

# used for univariate analysis
# used to find out the distribution 
# range of the observation
# central tendency 
# is the data bimodal
# are there outlires 


## Plot under Distribution Plot ##
# Histplot
# Kdeplot
# rugplot
```

```python
## figure level --> displot
## axes level --> histplot --> kdeplot --> rugplot 
```

```python
## plotting univariate histogram ##

sns.histplot(data=tips,x="total_bill")
```

```python
sns.displot(data=tips,x="total_bill",kind="hist")
```

```python
## bins parameter 

sns.displot(data=tips,x="total_bill",kind="hist",bins=20)
```

```python
## its also possible to visualize the distribution of a categorical variable using the logic of a histogram 
## Discrete bins are automatically set for the categorical variables 

tips 

sns.displot(data=tips,x="day",kind="hist")
```

```python
# hue parameter #

sns.displot(data=tips,x="tip",kind="hist",hue='sex')
```

```python
## element --> step 

sns.displot(data=tips,x="tip",kind="hist",hue='sex',element="step")
```

```python
titanic = sns.load_dataset('titanic')
```

```python
sns.displot(data=titanic,x="age",kind="hist",element='step',hue="sex")
```

```python
## faceting using col and row ----> not working on histplot function 

sns.displot(data=tips,x="tip",kind="hist",col='sex',element="step")
```

```python
## kdeplot ## kernel density estimate 

## rather than using discreate bins, a KDE plot smooths the observations with a gaussian kernel, producing a continuous density estimate 

tips

sns.kdeplot(data=tips,x='total_bill')

# kdplot probility batata hain jaise ki 20 ka hone ka probility hain 0.05 percentage hain 
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')

```

```python
## hue --> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue="sex",fill=True)
```

```python
## Rugplot 

# plot marginal distributions by drawing ticks alongthe x and y axes 

# this function is intended to complement other plots by showing the locations of individuals observations in an unobstrusive way 

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
## byviraient histogram 

sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='hist',height=10)  ## height matlab size badhana ficute level function main 
```
