When a python program runs all data such as variable list etc lives in a ram and when the program ends the data disapear forever
file handling solve this problem

```python
# Mode 'r' : Opens a file for reading only.

# "r" --> READ MODE 
# file must exsist, other wise you get fileNotFoundError

with open("data.text","r") as f:
    content=f.read()
    print(content)

```

```python
## safe version with error handling :

try:
    with open("file.text","r") as f:
        result=f.read()
        print(result)
except FileNotFoundError:
    print("File Does Not Exist")
```

## mode w - write(overwrite)
Open a file for writing if file does not exist Python Create it. if it does exist all previous content is erase and replace .Be carefull!

```python
with open("couse.txt","w") as f:
    f.write("Alice 95\n")
    f.write("Bob 87\n")
    
```

```python
## Mode "w" ---> Write(over)

with open("couse.txt","w") as f:
    f.write("Charlie 92\n")
    
```

# "a" -- Append Mode 

Opens a file to writing but move the cursor to the END of the existing content .New data is data after old data Existing content is never erased . Creates the file if it doest'n exist.

```python
# 'a' mode = Append (adds to the end, Never Deletes)

with open('Attendence.txt',"a") as f:
    f.write("Day 1: Alice,Bob,Charlie\n")
```

```python
with open("Attendence.txt","a") as f:
    f.write("Day 2 Alice,Daina,Eve\n")
```

```python
with open("Attendence.txt","a") as f:
    f.write("Day 3: Bob, Charlie,Eve\n")
```

```python
### "x" Create mode

try:
    with open("new.txt","x") as f:
        f.write("This is a brand new files\n")
        print("File Created Succesfully")
except FileExistsError:
    print("File is already Exist")
```
