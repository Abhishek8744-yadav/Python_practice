```python
def new_func(value):
    def wrapper():
        print("Good Morning")
        value ()
        print("Good Evining")
    return wrapper
```

```python
import time 

def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        print(start)
        func(*args,**kwargs)
        print(time.time())
        print("Time taken by this function is", time.time()-start)
     
    return wrapper
```

```python
@timer
def list_parcel(list):
    print(list)
    for i in list:
        print(i)
```

```python
list_parcel([1,2,4,45,54,12,21])
```

```python
import time
def code_timer(values):
    def wrapper(*args,**kwargs):
        start=time.time()
        print(f"Code start time is {start} sec")
        values(*args,**kwargs)
        end=time.time()
        print(f"Code End Time {end} sec")
        print()
        print()
        print(f"Total time taken to excute for the code is:- {end - start} ")
    return wrapper

```

```python

@code_timer
def odd_even(list):
    for i in list:
        if i%2==0:
            print(f"{i}:- Even")
        else:
            print(f"{i}:- odd")
```

```python
odd_even([10,20,4,7,56,54,4,6564])
```

```python
## basic syntax of Decorator ##

def decorator_name(orginal_name):
    def wrapper(*args,**kwargs):
        ## code before function runs
        result=orginal_function(*args,**kwargs)
        ## code after function runs
        return result
    return wrapper
        
```

```python
## how to call decoraters ##
@decorator_name 
def my_function1():
    print("Hello")
```

```python
## another way using decorater function ##

def my_function1():
    print("Hello")

my_function =decorator_name(my_function1)

my_function()
```

```python
## logging decirators(very common)

def log_decorator(func):
    def wrapper(*args,**kwargs):
        print(f"calling function: {func.__name__}")
        print()
        result=func(*args,**kwargs)
        print()
        print(f"Finished function:{func.__name__}")
        return result
    return wrapper
```

```python
@log_decorator
def greet(name):
    print(f"Hello {name}")

    
```

```python
greet("Abhishek")
```

```python
@log_decorator
def greet_name(name,age):
    print(f"Hello {name} you are {age} years old")
```

```python
greet_name("Abhishek",21)
```

```python
import time

def count_time_function(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        print(f"Start Time:- {start}")
        result=func(*args,**kwargs)
        print(f"The Time Taken To extute the code is {time.time()-start}")
        return result
    return wrapper
```

```python
@count_time_function
def count_num(num):
    for i in num:
        print(i)
```

```python
count_num([40,50,60,70])
```
