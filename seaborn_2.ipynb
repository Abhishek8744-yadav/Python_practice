```python
import seaborn as sns
```

```python
## import dataset 

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
sns.pairplot(iris)  ## histogram tab plot hoga jab same data ko same data k sath compare kiya jata hain aur ye hamesha daigonaly hi ata hain 
```

```python
## plotting pairwise Relationship (pairGrid vs Pairplot)
```

```python
## pair grid 

g = sns.PairGrid(data=iris)

## g.map

g.map(sns.scatterplot)
```

```python
## pair grid with hue parameter 

g = sns.PairGrid(data=iris,hue='species')

## g.map

g.map(sns.scatterplot)
```

```python
## mapdig --> map_offdiag 

g = sns.PairGrid(data=iris)

g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

```

```python
## map _diag ----> map_upper --> map_lower 
g = sns.PairGrid(data=iris)

g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
## vars isme aap bata sakte hain ki aap konse aur kitne column main kaam karna chahate ho 

g = sns.PairGrid(data=iris,vars=["sepal_width",'petal_width'])

g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
## jointGrid vs Jointplot ## 

sns.jointplot(data=tips,x="total_bill",y="tip",kind='hist',hue="sex")
```

```python
## joingrid ##

g = sns.JointGrid(data=tips,x="total_bill",y="tip")

g.plot(sns.scatterplot,sns.histplot)
```
