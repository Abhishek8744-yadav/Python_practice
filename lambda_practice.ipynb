```python
## Sort base on Accuracy with desc order ##

models= [
    {"name" : "RandomForest","accuracy":0.89},
    {"names":"XGBoost","accuracy":0.92},
    {"names":"SVM","accuracy":0.85}
]


sort_models=sorted(models,key=lambda x:x["accuracy"],reverse=True)

print(sort_models)
```

```python
## Decoraters ##

# A decorater is function that take function as input 

# Add extra functionality 

# returns a modified function 

# without changing the orginal function code 
```

```python
# This is original code 

def greet():
    print("Hello")
```

```python
greet()
```

```python
## basic decorater structure #

def decorater_function(greet_function):

    def wrapper():
        print("Good Morning!")
        greet_function()
        print("How Are You!")
    return wrapper
```

```python
decorated=decorater_function(greet)
```

```python
decorated()
```
